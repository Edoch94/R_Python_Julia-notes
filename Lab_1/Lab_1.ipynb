```python
import numpy as np
import matplotlib.pyplot as plt

beta = 1
alpha = 1
N = 400

r = np.random.rand(N)
t = (-(beta**(-1))*np.log(1-r))**(alpha**(-1))

delta_channel = .1
channel_centers = np.arange(delta_channel/2, 5, delta_channel)

num_samples = plt.hist(t, channel_centers)
pdf_est = num_samples[0]/(N*delta_channel)

analytic_weibull= alpha*beta*channel_centers**(alpha-1)*np.exp(-beta*channel_centers**alpha) 

plt.figure()
plt.plot(channel_centers, analytic_weibull, label='Analytic pdf')
plt.plot(channel_centers[:-1],pdf_est,'-sr', label='Sampled values distribution pdf')
plt.legend()

plt.figure() 
cdf_est= np.cumsum(pdf_est)*delta_channel
plt.plot(channel_centers, np.cumsum(analytic_weibull)*delta_channel, label='Analyticcdf')
plt.plot(channel_centers[:-1], cdf_est,'*r', label='Sampledvaluesdistributioncdf')
plt.legend()
```
